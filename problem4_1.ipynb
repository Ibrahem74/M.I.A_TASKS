```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather = pd.read_csv("Weather Dataset.csv")
weather
```

```python
weather.isnull() #no null values 
```

```python
weather.describe() # problem in temp humidity visibility loud cover (min value)

```

```python
weather = weather[weather['Temperature (C)'] != -2000]
weather = weather[weather['Humidity'] != -2000] 
weather = weather[weather['Visibility (km)'] != -2000]
weather = weather[weather['Loud Cover'] != -2000]
weather.describe()
```

```python
weather.duplicated() #there are duplicates 
```

```python
weather[weather['Formatted Date'].duplicated()==1] #showing duplicate data
```

```python
weather =weather.drop_duplicates()
weather.duplicated() # no duplicates
```

```python
weather=weather.drop(['Loud Cover'],axis=1) #not important zeros
weather
```

```python
weather.dtypes #all data types are correct
```

```python
plt.figure(figsize=(14,7))
time = weather.index
temperature= weather['Temperature (C)']
plt.xlabel('Time')
plt.ylabel('Temprature')
plt.title("Temperature over time")
plt.xticks([])
plt.plot(time,temperature,color='purple')
```

```python
plt.figure(figsize=(14,7))
plt.hist(temperature)
plt.ylabel('Temperature')
plt.xlabel('Time')
plt.xticks([])
```

```python
plt.figure(figsize=(14,7))
humidity = weather['Humidity']
plt.scatter(humidity,temperature)
plt.ylabel('Temperature')
plt.xlabel('Humidity')
plt.title("Temprature over Humidity")
```

```python
corr = weather.select_dtypes(include=['int64','float64']).corr()
corr
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(corr,cmap='coolwarm',annot=True)
plt.title("Correlation Heatmap")
```

Correlation shows the relation ship between two variables 

1 if there is a perfect direct relationship between the two virables
0 if there isn't any relationship between the two virables
-1 if there is a perfect inverse relationship between the two virables
